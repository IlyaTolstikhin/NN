```
from experta import*

class KnowledgeBase(KnowledgeEngine):
  def __init__(self):
    super().__init__()
    self.analyst = 0
    self.developer = 0
    self.WEB = 0
    self.SecAdmin = 0
    self.Nanalyst = 0
    self.Ndeveloper = 0
    self.NWEB = 0
    self.NSecAdmin = 0
  
  @Rule(
      Fact(answ="1-1") | Fact(answ="2-1") | Fact(answ="3-1") | Fact(answ="4-1"))
  def __(self):
    self.Nanalyst = self.Nanalyst + 0.15 * (1 - self.Nanalyst)
    self.developer = self.developer + 0.5 * (1 - self.developer)
    self.NWEB = self.NWEB + 0.1 * (1 - self.NWEB)
    self.NSecAdmin = self.NSecAdmin + 0.1 * (1 - self.NSecAdmin)

  @Rule(
      Fact(answ="1-2") | Fact(answ="2-2") | Fact(answ="4-2"))
  def plus2(self):
    self.analyst = self.analyst + 0.3 * (1 - self.analyst)
    self.Ndeveloper = self.Ndeveloper + 0.2 * (1 - self.Ndeveloper)
    self.NWEB = self.NWEB + 0.1 * (1 - self.NWEB)
    self.NSecAdmin = self.NSecAdmin + 0.15 * (1 - self.NSecAdmin)

  @Rule(
      Fact(answ="3-2"))
  def plus3(self):
    self.analyst = self.analyst + 0.45 * (1 - self.analyst)
    self.Ndeveloper = self.Ndeveloper + 0.15 * (1 - self.Ndeveloper)
    self.NWEB = self.NWEB + 0.2 * (1 - self.NWEB)
    self.SecAdmin = self.SecAdmin + 0.35 * (1 - self.SecAdmin)

  @Rule(
      Fact(answ="2-3"))
  def plus4(self):
    self.Nanalyst = self.Nanalyst + 0.15 * (1 - self.Nanalyst)
    self.Ndeveloper = self.Ndeveloper + 0.1 * (1 - self.Ndeveloper)
    self.WEB = self.WEB + 0.7 * (1 - self.WEB)
    self.SecAdmin = self.SecAdmin + 0.35 * (1 - self.SecAdmin)

  @Rule(
      Fact(answ="1-3") | Fact(answ="3-3") | Fact(answ="4-3"))
  def plus5(self):
    self.Nanalyst = self.Nanalyst + 0.1 * (1 - self.Nanalyst)
    self.Ndeveloper = self.Ndeveloper + 0.15 * (1 - self.Ndeveloper)
    self.WEB = self.WEB + 0.5 * (1 - self.WEB)
    self.NSecAdmin = self.NSecAdmin + 0.2 * (1 - self.NSecAdmin)

  @Rule(
      Fact(answ="1-4") | Fact(answ="4-4"))
  def plus6(self):
    self.Nanalyst = self.Nanalyst + 0.1 * (1 - self.Nanalyst)
    self.Ndeveloper = self.Ndeveloper + 0.2 * (1 - self.Ndeveloper)
    self.NWEB = self.WEB + 0.25 * (1 - self.NWEB)
    self.SecAdmin = self.SecAdmin + 0.45 * (1 - self.SecAdmin)

  def getKU(self):
    return {
        "Аналитик": self.analyst - self.Nanalyst,
        "Разработчик ПО": self.developer - self.Ndeveloper,
        "Веб-разработчик":  self.WEB -  self.NWEB,
        "Информационная безопасность или Системный администратор": self.SecAdmin - self.NSecAdmin
          }
                      
  def checkOS(self):
    ku = self.getKU()
    max = -1
    key = ""
    for res in ku:
      if ku[res] > max:
        max = ku[res]
        key = res

    return key, max 
  
```

```
KBase = KnowledgeBase()
KBase.reset()
for ThisKey in Questons:
    print(Questons[ThisKey])
    x = str(input(Answer[ThisKey]))
    ans = str(ThisKey) + "-" + x
    KBase.declare(Fact(answ=ans))
    KBase.run()

check = KBase.checkOS()

print("Вы приняты на работу на должность: " + check[0])
```
